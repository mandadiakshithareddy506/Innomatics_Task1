```python
username = "admin"
password = "1234"
input_username = input("Enter username: ") # user inputs
input_password = input("Enter password: ") # user inputs
if input_username == username and input_password == password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
marks = [45, 78, 90, 33, 60] # given marks
pass_count = 0
fail_count = 0
for mark in marks: # conditional statement  for checking each student marks
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

print("Marks:", marks) # print final results
print("Total Students:", len(marks))
print("Total Pass Students:", pass_count)
print("Total Fail Students:", fail_count)
```

```python
names = [" Alice ", "bob", " CHARLIE "] # given names
cleaned_names = []

for name in names:
    cleaned_name = name.strip().lower()
    cleaned_names.append(cleaned_name)

print("Original Names:", names)
print("Cleaned Names:", cleaned_names)
```

```python
messages = ["Hi", "Welcome to the platform", "OK"]  # given messages

for message in messages:
    length = len(message)
    print("Message:", message)
    print("Length:", length)

    if length > 10:
        print("Status: Longer than 10 characters")
    else:
        print("Status: Within limit")

    print()
```

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"] # given logs

error_count = 0

for log in logs: # condition statement for checking each logs entry
    if log == "ERROR":
        error_count += 1

print("Total ERROR messages:", error_count)
```
